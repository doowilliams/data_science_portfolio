```python
x = df.iloc[:,1:5]
print(x)
y = df.iloc[:,5]
print(y)
```

Creating KNN Model


Creating testing and training datasets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, train_size=0.8, random_state=88, shuffle=True)
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=6, metric='minkowski', p=1)

KNN.fit(x_train, y_train)

KNN_XX = np.array([[0.1,1.5,6,2]])

KNN.predict(KNN_XX)
```

**Evaluating our model using Score Accuracy**

```python
predict_type = KNN.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict_type)
```
